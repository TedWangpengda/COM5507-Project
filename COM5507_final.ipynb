```python
import yaml
import os
import requests
import time
import random
import openpyxl
```

```python
def geturl(fakeid, rootpath):
    history_file = os.path.join(rootpath, "history.txt")
    history = set()

    if os.path.exists(history_file):
        with open(history_file, "r") as file:
            for line in file:
                url = line.strip()
                history.add(url)

    return history
```

```python
def getwxlist(fakeid,rootpath):
    with open("wechat.yaml", "r",encoding=('utf-8')) as file:
        file_data = file.read()
    config = yaml.safe_load(file_data)
    headers = {
        "Cookie": config['cookie'],
        "User-Agent": config['user_agent'] 
        }
    # 请求参数
    url = "https://mp.weixin.qq.com/cgi-bin/appmsg"
    begin = "0"
    params = {
        "action": "list_ex",
        "begin": begin,
        "count": "5",
        "fakeid": "MjM5MjU0NTc4OQ%3D%3D",
        "type": "9",
        "token": config['token'],
        "lang": "zh_CN",
        "f": "json",
        "ajax": "1"
    }
    wxlistfile = os.path.join(rootpath, "wxlist.xlsx")
    #获取历史抓取记录
    history = geturl(fakeid,rootpath)
    wb=openpyxl.load_workbook(wxlistfile)
    ws = wb.active 
    # 在不知道公众号有多少文章的情况下，使用while语句,也方便重新运行时设置页数
    i = 0
    #新增抓取计数
    j = 0
    flag = False
    while True:
        begin = i * 5
        params["begin"] = str(begin)
        # 随机暂停几秒，避免过快的请求导致过快被查到
        time.sleep(random.randint(1,10))
        resp = requests.get(url, headers=headers, params = params, verify=False)
        # 微信流量控制, 退出
        if resp.json()['base_resp']['ret'] == 200013:
            print("frequencey control, stop at {}".format(str(begin)))
            time.sleep(3600)
            continue
            # 如果返回的内容中为空则结束
        if len(resp.json()['app_msg_list']) == 0:
            print("all ariticle parsed")
            break
        msg = resp.json()
        if "app_msg_list" in msg:
            for item in msg["app_msg_list"]:
                if item['link'] in history:
                    flag = True
                    break 
                row = [item['title'], item['link']] 
                ws.append(row)
                   
                j += 1
            print(f"第{i}页爬取成功\n")
        if flag == True:
            print(f"新增页面抓取已完成,{j}篇文章已添加\n")
            break
        # 翻页
        i += 1  
    wb.save(wxlistfile)
```

```python

tokenized_texts = []

for title in df1['Title']:
    if title and isinstance(title, str):
        tokens = jieba.cut(title, cut_all=False)
        tokenized_text = ' '.join(tokens)
        tokenized_texts.append(tokenized_text)
    else:
        print("Skipping a non-string or empty title:", title)

# Load the custom stopwords file and combine with additional stopwords
with open("哈工大停用词表.txt", "r", encoding="utf-8") as file:
    chinese_stopwords = set(file.read().splitlines()) | set(stopword_wechat) | set(stopword_douyin)

# 删除停用词
filtered_texts = []
for text in tokenized_texts:
    words = text.split()
    filtered_text = ' '.join([word for word in words if word not in chinese_stopwords])
    filtered_texts.append(filtered_text)

# Print filtered texts
for text in filtered_texts:
    print("Filtered Text:", text)

# 更新 filtered_tokens 变量
filtered_tokens = []
for text in filtered_texts:
    tokens = text.split('/')
    for token in tokens:
        if token not in chinese_stopwords:
            filtered_tokens.append(token)

# 打印分词后的结果
print("分词后:", "/".join(filtered_tokens))

# 使用TF-IDF进行特征提取
tfidf_vectorizer = TfidfVectorizer(max_df=0.95, min_df=2, max_features=1000)
tfidf = tfidf_vectorizer.fit_transform(filtered_tokens)


# ----------  Topic Number Choice  ----------- #
# Function to compute coherence values
def compute_coherence_values(dictionary, corpus, texts, limit, start=2, step=3):
    coherence_values = []
    model_list = []
    for num_topics in range(start, limit, step):
        # Build LDA model
        model = gensim.models.ldamodel.LdaModel(corpus=corpus, id2word=dictionary, num_topics=num_topics)
        model_list.append(model)
        coherencemodel = CoherenceModel(model=model, texts=texts, dictionary=dictionary, coherence='c_v')
        coherence_values.append(coherencemodel.get_coherence())

    return model_list, coherence_values

# Create Dictionary
id2word = gensim.corpora.Dictionary([text.split() for text in filtered_texts])

# Create Corpus: Term Document Frequency
corpus = [id2word.doc2bow(text.split()) for text in filtered_texts]

# Define the range for the number of topics
start, limit, step = 2,14, 1

# Compute coherence scores for different numbers of topics
model_list, coherence_values = compute_coherence_values(dictionary=id2word, corpus=corpus, texts=[text.split() for text in filtered_texts], start=start, limit=limit, step=step)
# Print the coherence scores
for num_topics, cv in zip(range(start, limit, step), coherence_values):
    print("Num Topics =", num_topics, " has Coherence Value of", round(cv, 4))

# Select the model and print the topics
optimal_model = model_list[np.argmax(coherence_values)]
optimal_model.print_topics(num_words=10)

plt.plot(range(start, limit, step), coherence_values)
plt.xlabel("Number of Topics")
plt.ylabel("Coherence score")
plt.title("Coherence Scores for Different Number of Topics")
# 保存图像为JPG文件
plt.savefig('Coherence_wechat.jpg', dpi=500)
plt.show()
```

```python

# ----------  Topic Results  ----------- #
# 查看主题并保存到表格
def display_topics(model, feature_names, no_top_words, perplexity):
    topics = []
    for topic_idx, topic in enumerate(model.components_):
        topic_words = [feature_names[i] for i in topic.argsort()[:-no_top_words - 1:-1]]
        topics.append(topic_words)

    df_topics = pd.DataFrame(topics, columns=[f"Word {i+1}" for i in range(no_top_words)])
    df_topics.index.name = 'Topic'
    df_topics['Model Perplexity'] = perplexity  # 添加模型困惑度列
    return df_topics

# 应用LDA主题模型
n_topics = 10  # the coherence score for 7 topics is the highest
lda = LatentDirichletAllocation(n_components=n_topics, max_iter=5, learning_method='online', learning_offset=50.,random_state=0)
lda.fit(tfidf)

# 获取特征词汇列表
feature_names = tfidf_vectorizer.get_feature_names_out()

# 计算模型困惑度
model_perplexity = lda.perplexity(tfidf)

# 调用函数并显示结果
no_top_words = 15
df_topics = display_topics(lda, feature_names, no_top_words, model_perplexity)
df_topics.to_excel('wechat_whole.xlsx')
print(df_topics)
```

```python
# ----------  Pre Set  ----------- #
import os
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import numpy as np

os.getcwd()
new_path = 'D:\\OneDrive - The Chinese University of Hong Kong\\2023Fall\\MediaData'
os.chdir(new_path)
```

```python
# ----------  Merge Data  ----------- #
# Import Wechat and Douyin Data
df_wechat = pd.read_excel('wechat_titlelist_all.xlsx', header = 0, sheet_name="Sheet2")
df_douyin = pd.read_excel('douyin_titlelist_all.xlsx', header = 0, sheet_name="Sheet2")

# Aggregate to Month (Max-Like)
Month_Max = df_douyin.groupby('Date')['Likes'].max().reset_index()
Month_Max['Month'] = pd.to_datetime(Month_Max['Date'], format='%Y-%m')

# Aggregate to Month (Total-Articles)
Month_Total = df_wechat.groupby('Month')['Title'].count().reset_index()
Month_Total['Month'] = pd.to_datetime(Month_Total['Month'], format='%Y-%m')

Month_Both = pd.merge(Month_Max, Month_Total, on = 'Month', how='outer')

# Import Economic Data
df_ML = pd.read_excel('Cycle.xlsx', header = 0, sheet_name="Sheet1")
df_CMI = pd.read_excel('Cycle.xlsx', header = 0, sheet_name="Sheet3")

# Extract Month Info
df_CMI['Date'] = pd.to_datetime(df_CMI['Date'])
df_ML['Time'] = pd.to_datetime(df_ML['Time'])

df_CMI['Month'] = df_CMI['Date'].dt.strftime('%Y-%m')
df_CMI['Month'] = pd.to_datetime(df_CMI['Month'], format='%Y-%m')
df_ML['Month'] = df_ML['Time'].dt.strftime('%Y-%m')
df_ML['Month'] = pd.to_datetime(df_ML['Month'], format='%Y-%m')

df_econ = pd.merge(df_ML, df_CMI, on="Month", how='outer')

# Merge
df = pd.merge(Month_Both, df_econ, on='Month', how = 'outer')

# Oder Adjust
new_order = ['Month', 'Title', 'Likes', 'Cycle_label', 'CMI', 'PMI', 'PPI']
df = df.reindex(columns=new_order)

# Rename
column_mapping = {'Month': 'Month', 
                  'Title': 'Article_Number', 
                  'Likes': 'Likes_Max', 
                  'Cycle_label':'Economic Situation', 
                  'CMI': 'CMI',
                  'PMI': 'PMI', 
                  'PPI':'PPI'}
df = df.rename(columns=column_mapping)

df.loc[df['Economic Situation'] == '衰退', 'Economic Situation'] = 'Reflation'
df.loc[df['Economic Situation'] == '滞胀', 'Economic Situation'] = 'Stagflation'
df.loc[df['Economic Situation'] == '复苏', 'Economic Situation'] = 'Recovery'
df.loc[df['Economic Situation'] == '过热', 'Economic Situation'] = 'Overheat'

print(df.head(10))
# export as excel
pd_merged = pd.DataFrame.from_dict(df)

pd_merged.to_excel('Merged_ECON.xlsx', index=False, encoding='utf-8')
```

```python
# ----------  Merrill Lynch investment clock  ----------- #
df = df.sort_values('Month')
df['Date End'] = df['Month'] + pd.Timedelta(days=30)

conditions = [(df['Economic Situation'] == 'Recovery'), 
              (df['Economic Situation'] == 'Overheat'), 
              (df['Economic Situation'] == 'Stagflation')]
choices = ['#FF0000', '#00FF00', '#0000FF']

df['Color'] = np.select(conditions, choices, default='gray')

fig, ax = plt.subplots(figsize=(10, 6))

for i, row in df.iterrows():
    ax.axvspan(row['Month'], row['Date End'], color=row['Color'], alpha=0.2)


# Time Series
ax.plot(df['Month'], df['Article_Number'], color='black')

# Title and Lable
ax.set_title('Number of Articles and Economic Situation')
ax.set_xlabel('Time')
ax.set_ylabel('Number of Articles')

# Legend
ax.plot([], [], color='#FF0000', alpha=0.2, label='Recovery')
ax.plot([], [], color='#00FF00', alpha=0.2, label='Overheat')
ax.plot([], [], color='#0000FF', alpha=0.2, label='Stagflation')
ax.plot([], [], color='gray', alpha=0.2, label='Reflation')

ax.legend()

# Save and Show
plt.savefig('Num_ML.jpg', dpi=500)
plt.show()
```

```python
# ----------  CMI & Likes  ----------- #
# Time Series
df.set_index('Month', inplace=True)
df = df[(df.index >= '2019-01-01')]

# Create
fig, ax1 = plt.subplots()

# First Line
ax1.plot(df.index, df['CMI'], linestyle='-', label='CMI')
ax1.set_ylabel('CMI')

# Second Line
ax2 = ax1.twinx()

ax2.plot(df.index, df['Likes_Max'], linestyle='-', color='r', label='Max Likes')
ax2.set_ylabel('Max Likes w/ Month')

# Title and Label
ax1.set_title('Trend: CMI and the Max Likes')
ax1.set_xlabel('Time')

# Legend
lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
ax1.legend(lines_1 + lines_2, labels_1 + labels_2)

fig.autofmt_xdate()

# Save and Show
plt.savefig('Trend_CMI_Likes.jpg', dpi=500)

plt.show()
```
